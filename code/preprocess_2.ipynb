## Code to preprocess the data

```python
from indicnlp.normalize.indic_normalize import GujaratiNormalizer
```

```python
# loading the gujarati normalizer
normalizer = GujaratiNormalizer()
```

```python
# loading the data
with open("../toy_data/gu/gu.sent.1m.orig") as f:
    lines = f.read()
    lines = [l for l in lines.split("\n") if l]
```

```python
words = []

for line in l:
    for wl in line.split(" "):
        if wl and wl not in words:
            words.append(wl)
```

```python
# Replacing the punctuations with space and normalizing the lines.
# Following steps followed by Ashwini et al. 
replaced_lines = []
for line in lines:
    line = line.translate(str.maketrans('', '', string.punctuation))
    line = normalizer.normalize(line)
    replaced_lines.append(line)
```

```python
with open("../toy_data/gu/gu.sent.preproc.1m", "w") as f:
    f.write("\n".join(replaced_lines))
#     lines = [l for l in lines.split("\n") if l]
```

```python
len(replaced_lines)
```

```python
tokens = []
for line in replaced_lines:
    for word in line.split(" "):
        word = word.strip()
        if word not in tokens:
            tokens.append(word)
```

```python
#checking the useful and non-useful words
useful = {}
not_useful = {}
useful_lines = []
count = 0
with open("../toy_data/gu/gu.sent.1m.dict") as f:
    file_lines = f.read()
#     file_lines = [f for f in file_lines.split("\n") if f]
    for line in file_lines.split("\n"):
        if line:
            word, word_count = line.split("\t")
            word_count = int(word_count)
            if int(word_count) >= 100:
                useful[word] = word_count
                useful_lines.append(line)
            else:
                not_useful[word] = word_count
        count += 1
        if count % 10000 == 0:
            print("{} done".format(count))
```
